```python
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
X = np.array([[0.5, 1.5], [1, 1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y = np.array([0, 0, 0, 1, 1, 1])
```

```python
model = LogisticRegression()
model.fit(X, y)
w_final = model.coef_
b_final = model.intercept_

print(f"model parameters:\nw: {w_final}, b:{b_final}")
```

```python
n = X.shape[1]

features = np.zeros(n)

for i in range(n):
    value = float(input(f"Enter x{i+1}: "))

    features[i] = value
```

```python
pred = model.predict(features.reshape(1, -1))  # makes it a 2D array of 1 x n
print(f"Prediction : {pred}")
```

```python
score = model.score(X, y)
print(f"Accuracy Score: {score}")
```
